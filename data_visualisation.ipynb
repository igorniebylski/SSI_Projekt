```python
import numpy as np
import pandas as pd
from matplotlib import pyplot as plt
from sklearn import decomposition
```

```python
data = pd.read_csv("numbers.csv")
labels = data.iloc[:, 0]
data = data / 255.
labels = labels.astype('int')
print(np.unique(labels))
```

```python
pca = decomposition.PCA(n_components=2)
view = pca.fit_transform(data)
plt.scatter(view[:,0], view[:,1], c=labels, alpha=0.2, cmap='Set1')
```

```python
from sklearn.manifold import TSNE
import random as r

n = r.randint(1, 15900)
view = TSNE(n_components=2, random_state=n).fit_transform(data)
plt.figure(figsize=(20,10))
plt.scatter(view[:,0], view[:,1], c=labels, alpha=0.5)
plt.xlabel('t-SNE-1')
plt.ylabel('t-SNE-2')
```
